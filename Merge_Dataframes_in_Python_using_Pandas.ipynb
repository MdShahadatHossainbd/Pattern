```
import pandas as pd 
```

```
df = pd.read_csv("data.csv")
```

```
df
```

```
df2 = pd.read_csv("dataset.csv")
df2
```

```
df3 = pd.merge(df,df2,on='ID')
```

```
df3
```

```
df3 = pd.merge(df,df2,on='ID',how='outer') #inner
df3
```

```
df3 = pd.merge(df,df2,on='ID',how='left')
df3
```

```
df3 = pd.merge(df,df2,on='ID',how='right')
df3
```

```
df3 = pd.merge(df,df2,on='ID',how='right',indicator=True)
df3
```
